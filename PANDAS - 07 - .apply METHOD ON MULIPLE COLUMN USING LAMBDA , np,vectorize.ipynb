```python
def quality(total_bill , tip):
    
    if tip / total_bill > 0.25:
        
        return "Generous"
    
    else:
        
        return "Other"
```

```python
quality(16.00 , 5.00) 
```

```python
data
```

```python
data["quality_of_tip"] = data[["total_bill" , "tip"]].apply(lambda  data : quality(data["total_bill"] , data["tip"]), axis = 1  )
```

### Instead of lambda use np.vectorize for shorter code

```python
data["quality_of_tip"] = np.vectorize(quality)(data["total_bill"] , data["tip"])
```
