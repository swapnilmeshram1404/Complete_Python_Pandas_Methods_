### check how pricy is the restraurant, if total_bill < 10 $,
totall bill > 10 upto 30 $$,  
above 30 $$$

```python
def pricy(num):
    
    if num < 10:
        
        return "$"
    
    elif num >10 and num<30:
        
        return "$$"
    
    else:
        
        return "$$$"
    
```

```python
data[" how pricy"] = data["total_bill"].apply(pricy)
```
